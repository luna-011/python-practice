```python
time_slots = ['9:00', '10:00', '11:00', '12:00', '1:00']

new_car_id = [
  ["18-04-2025", "Sunday",    [6, 10, 1, 10, 7]],
  ["19-04-2025", "Monday",    [10, 2, 10, 6, 4]],
  ["20-04-2025", "Tuesday",   [3, 1, 10, 10, 9]],
  ["21-04-2025", "Wednesday", [10, 8, 10, 7, 3]],
  ["22-04-2025", "Thursday",  [2, 10, 4, 10, 1]]
]

renew_car_id = [
  ["18-04-2025", "Sunday",    [9, 8, 7, 10, 10]],
  ["19-04-2025", "Monday",    [2, 9, 10, 3, 10]],
  ["20-04-2025", "Tuesday",   [1, 1, 10, 9, 10]],
  ["21-04-2025", "Wednesday", [10, 3, 9, 10, 9]],
  ["22-04-2025", "Thursday",  [10, 5, 2, 10, 3]]
]

new_driving_license = [
  ["18-04-2025", "Sunday",    [10, 2, 10, 4, 1]],
  ["19-04-2025", "Monday",    [3, 10, 4, 10, 5]],
  ["20-04-2025", "Tuesday",   [9, 10, 2, 6, 10]],
  ["21-04-2025", "Wednesday", [10, 3, 10, 5, 7]],
  ["22-04-2025", "Thursday",  [2, 10, 4, 6, 10]]
]

renew_driving_license = [
  ["18-04-2025", "Sunday",    [10, 4, 5, 10, 1]],
  ["19-04-2025", "Monday",    [5, 6, 10, 3, 10]],
  ["20-04-2025", "Tuesday",   [10, 8, 1, 3, 10]],
  ["21-04-2025", "Wednesday", [2, 9, 10, 10, 6]],
  ["22-04-2025", "Thursday",  [10, 7, 4, 10, 3]]
]

lost_id = [
  ["18-04-2025", "Sunday",    [9, 10, 10, 9, 1]],
  ["19-04-2025", "Monday",    [3, 10, 4, 10, 4]],
  ["20-04-2025", "Tuesday",   [2, 9, 6, 10, 10]],
  ["21-04-2025", "Wednesday", [6, 10, 5, 10, 2]],
  ["22-04-2025", "Thursday",  [9, 10, 4, 10, 4]]
]

services_data = {
  1: ("اصدار رخصة سيارة اول مرة", new_car_id),
  2: ("تجديد رخصة سيارة", renew_car_id),
  3: ("اصدار رخصة قيادة اول مرة", new_driving_license),
  4: ("تجديد رخصة قيادة", renew_driving_license),
  5: ("اصدار بدل فاقد", lost_id)
}

services = {
    1: "اصدار رخصة سيارة اول مرة",
    2: "تجديد رخصة سيارة",
    3: "اصدار رخصة قيادة اول مرة",
    4: "تجديد رخصة قيادة",
    5: "اصدار بدل فاقد"
  }

def booking_system():
  print("\n--- Book a Service ---")
  for num, name in services_data.items():
    print(f"{num}. {name[0]}")

  while True:
    try:
      service_choice = int(input("Choose a service by number: "))
      if service_choice in services_data:
        break
      else:
        print("Invalid choice.")
    except:
      print("Please enter a valid number.")

  selected_service_name, selected_data = services_data[service_choice]

  print("\nAvailable Dates:")
  for i, (date, day, slots) in enumerate(selected_data):
    print(f"{i + 1}. {day} {date}")

  while True:
    try:
      date_choice = int(input("Choose a date by number: ")) - 1
      if 0 <= date_choice < len(selected_data):
        break
      else:
        print("Invalid choice.")
    except:
      print("Please enter a valid number.")

  date, day, slots = selected_data[date_choice]

  print("\nAvailable Time Slots:")
  for i, (time, count) in enumerate(zip(time_slots, slots)):
    status = "Full" if count >= 10 else f"{10 - count} left"
    print(f"{time}: {status}")

  while True:
    slot_choice = input("Choose a time slot: ")
    if slot_choice in time_slots:
      slot_index = time_slots.index(slot_choice)
      if slots[slot_index] >= 10:
        print("That slot is full. Choose another.")
      else:
        print("Slot selected successfully!\n")
        break
    else:
      print("Invalid time slot. Try again.")

  full_name = input("Enter your full name: ")
  user_id = input("Enter your ID: ")
  selected_data[date_choice][2][slot_index] += 1

  print("\n--- Reservation Confirmed ---")
  print(f"Name: {full_name}")
  print(f"ID: {user_id}")
  print(f"Service: {selected_service_name}")
  print(f"Date: {date} ({day})")
  print(f"Time: {time_slots[slot_index]}")

def main():
  print("Welcome to BMS (Booking Management System)")

  booking_system()

main()
```
